```python
def bestSupervisedMethod(model_stg, X, y, parameters={}, k_cv=5, rs=0, n_jobs=1):
    '''
    Return best hyperparameter model. 

    Parameters
    ----------
    model : sklearn
        Scikit-Learn method.
    parameters : dict
        Dictionary of parameters.
    X : array
        Features data.
    y : array
        Response data.
    k_cv : int, optional
        Cross validation value. The default is 5.
    rs: int, optional
        Random State

    Returns
    -------
    Dcictionary composed of the more suitable hyperparameters for the method.
    And the grid search Cv results.

    '''
    model = eval(model_stg)
    grid = GridSearchCV(
        estimator=model(),
        param_grid=parameters,
        n_jobs=n_jobs,
        scoring='neg_mean_squared_error')
    grid.fit(X, y)
    ind_prm = grid.cv_results_['rank_test_score'][0] -1
    

    scores = cross_validate(
        estimator= model(**grid.cv_results_['params'][ind_prm]),
        X=X, y=y, n_jobs=n_jobs,
        scoring=('r2', 'neg_mean_squared_error'),
        n_jobs=n_jobs,
        cv=KFold(n_splits=k_cv, shuffle=True, random_state=rs))

    text = f'For {model} we have MSE:'+\
          f"{int(scores['test_neg_mean_squared_error'].mean())} +-"+\
          f"{int(scores['test_neg_mean_squared_error'].std())}"+\
          f'\nFor {model} we have R^2:'+\
          f"{round(100*scores['test_r2'].mean(), 1)} +-"+\
          f"{round(100*scores['test_r2'].std(), 1)}"
    print(text)
    return {'bestHyperParams':grid.cv_results_['params'][ind_prm],
            'gridSearch': grid.cv_results_, 'recap':text}

def bestSupervisedMethod_pip(dct, X, y, prms={}, k_cv=5, rs=0, n_jobs=1):
    '''
    Return best hyperparameter model for a pipeline estimator. 

    Parameters
    ----------
    model : sklearn
        Scikit-Learn method.
    parameters : dict
        Dictionary of parameters.
    X : array
        Features data.
    y : array
        Response data.
    k_cv : int, optional
        Cross validation value. The default is 5.
    rs: int, optional
        Random State. The default is 0.
    n_jobs : int, optional
        Core number to use. The default is 1.

    Returns
    -------
    Dcictionary composed of the more suitable hyperparameters for the method.
    And the grid search Cv results.

    '''
    dct1 = {k: eval(dct[k]) for k in dct}
    if not prms:
        pip_estim=[]
    else:
        pip_estim = [(k, dct1[k]()) for k in dct1]
    
    grid = GridSearchCV(
        estimator=Pipeline(pip_estim),
        param_grid=prms,
        n_jobs=n_jobs,
        scoring='neg_mean_squared_error',
        cv = KFold(n_splits=k_cv, shuffle=True, random_state=rs))
    
    grid.fit(X, y)
    ind_prm = grid.cv_results_['rank_test_score'][0] -1
    
    dct_2 = dct1.copy()
    for k in prms:
        print(k.split('__')[0])
        dct_2[k.split('__')[0]] =\
            dct1[k.split('__')[0]](
                **{k.split('__')[1] : grid.cv_results_['params'][ind_prm][k]})
    pip_estim_2 = [(k, dct_2[k]) for k in dct_2]        
    scores = cross_validate(
        estimator= Pipeline(pip_estim_2),
        X=X, y=y, n_jobs=n_jobs,
        scoring=('r2', 'neg_mean_squared_error'),
        n_jobs=n_jobs,
        cv=KFold(n_splits=k_cv, shuffle=True, random_state=rs))

    text = f'For {Pipeline(pip_estim)} we have MSE:'+\
          f"{int(scores['test_neg_mean_squared_error'].mean())} +-"+\
          f"{int(scores['test_neg_mean_squared_error'].std())}"+\
          f'\nFor {Pipeline(pip_estim)} we have R^2:'+\
          f"{round(100*scores['test_r2'].mean(), 1)} +-"+\
          f"{round(100*scores['test_r2'].std(), 1)}"
    print(text)
    return {'bestHyperParams':grid.cv_results_['params'][ind_prm],
            'gridSearch': grid.cv_results_, 'recap':text}

def compute_fct_num(X, y, col_ind, estim_stg, score=('neg_mean_squared_error', 'r2'), 
                prms={}, k_cv=5, rs=0, n_jobs=1):
    '''

    Return best hyperparameter model for a pipeline estimator. 

    Parameters
    ----------
    X : array
        Features data.
    y : array
        Outcome data.
    col_ind : list/array
        Column index to select in X.
    estim_stg : str
        Estimator.
    score : tuple, optional
        Tuple of the name to use  score metrics. The default is ('neg_mean_squared_error', 'r2').
    prms : set, optional
        Set having as keys estimator argument and values the values of these arguements. The default is {}.
    k_cv : int, optional
        Cross validation value. The default is 5.
    rs: int, optional
        Random State. The default is 0.
    n_jobs : int, optional
        Core number to use. The default is 1.

    Returns
    -------
    TYPE
        DESCRIPTION.

    '''
    X_m = X.iloc[:, col_ind]
    crv = cross_validate(
        estimator = eval(estim_stg)(**prms),
        X = X_m.values.reshape(X_m.shape[0],-1),
        y=y.values,
        scoring=score,
        cv=KFold(n_splits=k_cv, shuffle=True, random_state=rs),
        n_jobs=n_jobs) 
    if type(score)==str:
        score = score.split(',')
    else:
        pass
    score_result = {k:[0, 0] for k in score}
    for i,k in enumerate(score):
        if k == 'r2':
            score_result[k][0] = round(100*crv['test_'+score[i]].mean(), 1)
            score_result[k][1] = round(100*crv['test_'+score[i]].std(), 1)
        else:
            score_result[k][0] = round(np.abs(crv['test_'+score[i]]).mean(), 1) 
            score_result[k][1] = round(crv['test_'+score[i]].std(), 1)
    return np.array([score_result[k] for k in score]).reshape(1,-1)[0]

def cv_forwardStepwiseSelection_puiss(df, estim_stg='linear_model.LinearRegression',
    score=('neg_mean_squared_error', 'r2'), prms={}, k_cv=5, rs=0, n_jobs=1,
    min_score = True):
    '''
    <cv_forwardStepwiseSelection> makes forward stepwise selection using OLS
    regression as model and relying on estimated MSE, provided by k-fold 
    cross-validation method.

    Parameters
    ----------
    df : DataFrame
        Outcome data.
    estim_stg : str
        Estimator.
    score : tuple, optional
        Tuple of the name to use  score metrics. The default is ('neg_mean_squared_error', 'r2').
    prms : set, optional
        Set having as keys estimator argument and values the values of these arguements. The default is {}.
    k_cv : int, optional
        Cross validation value. The default is 5.
    rs: int, optional
        Random State. The default is 0.
    n_jobs : int, optional
        Core number to use. The default is 1.
    min_score : boolean, optional.
        True if we want to minimize <score[0]> otherwise to maximize it. The default is True

    Returns
    -------
    list
        Assume that the the data frame contains p features, the first item
        will be a data frame containing p + (p-1) + (p-2) + ... + 1 = p(p+1)/2
        rows. The first p rows correspond to the models containing only one 
        feature, next among these first p model we choose this one having the
        lower estimated mse. Then we add to the later model an other feature 
        among the remaining p-1 features and we keep the model having the lower
        estimated mse and so on.
        
        The second item is a data frame is summary of the best models of the 
        first item. It contains p rows, that each one correspond to the best 
        model, that have the lower estimated MSE for a fixed feature number.
        
        For the 2 items the data are organized in 3 columns like below:
        
        +--------------------+---------------------+------------------------+
        | number of features | indexes of features | estimated mse (k-fold) |
        +--------------------+---------------------+------------------------+
        |         3          |     (x8, x3, x6)    |      1587              |
        +--------------------+---------------------+------------------------+
        
    '''
    
    dur = np.array(2 * [-time.time()])
    # a. Initialization
    cols = np.array([[k+'_mean', k+'_std'] for k in score]).reshape(1,-1)[0]
    y = df.iloc[:, 0]
    X = df.iloc[:, 1:]
    p = X.shape[1]
    df_fwd_list = []
    df_fwd_list.append(pd.DataFrame({"predictorNumb":0,
                                     "combinations":[tuple()],
                                     "model": estim_stg
                                     })) 
    y_m = y.mean()
    y_mean = pd.DataFrame({'mean': [y_m for k in y]})                  
    s = df_fwd_list[-1].loc[:, ['combinations', 'model']].apply(
            lambda x: compute_fct_num(X=y_mean, y=y, col_ind=0, estim_stg=x[1],
                prms=prms, score=score, cv=cv, n_jobs=n_jobs), axis=1)                  
    for k in cols:
        df_fwd_list[-1].loc[:, k] = 0
    df_fwd_list[-1].loc[:, cols] = pd.DataFrame(s.tolist(), columns=cols, index=s.index)
    score_up = score[0]+'_up'   
    df_fwd_list[-1].loc[:, score_up] = df_fwd_list[-1].loc[:, cols[0]]  +\
        df_fwd_list[-1].loc[:, cols[1]]    
    model_fwd_list = df_fwd_list.copy()
    ind_tuple = tuple() 
    ind_tuple_vs = tuple([k for k in np.arange(p) if not k in ind_tuple])
    # b. Looping
    for j in range(p):
        dur_i =  np.array(2 * [-time.time()])                                   # array to compute duration.
        indexT = np.array([list(ind_tuple)+[k] for k in ind_tuple_vs])          
        df_f = pd.DataFrame({
            "predictorNumb": np.repeat(j+1, len(indexT)),        
            "combinations": [tuple(x) for x in indexT],
            "model": [estim_stg for k in indexT]
            })
        s = df_f.loc[:, ['combinations', 'model']].apply(
            lambda x: compute_fct_num(X=X, y=y, col_ind=list(x[0]), estim_stg=x[1],
                prms=prms, score=score, cv=cv, n_jobs=n_jobs), axis=1)                  
        for k in cols:
            df_f.loc[:, k] = 0
        df_f.loc[:, cols] = pd.DataFrame(s.tolist(), columns=cols, index=s.index)
        score_up = score[0]+'_up'
        df_f[score_up] = df_f[cols[0]]  + df_f[cols[1]]
        print(f"Iteration number: {j}\nColumn number: {indexT.shape[1]}")       
        df_fwd_list.append(df_f)
        model_fwd_list.append(
            df_f.sort_values(by=score_up, ascending=min_score).iloc[[0], :])
        ind_tuple += tuple([model_fwd_list[-1].combinations.values[0][-1]])
        ind_tuple_vs = tuple([k for k in np.arange(p) if not k in ind_tuple])
        dur_i[1] = time.time()
        qh, rh = dur_i.sum()// 3600, dur_i.sum() % 3600
        qm, rm = rh // 60, rh % 60
        print(f"Model predictor model treated:{j}/{p-1}",\
              f"\nElapsed time: {int(qh)}h {int(qm)}min {int(rm)}s")
    # c. Adding
    df_fwd_total = pd.concat(df_fwd_list).reset_index(drop=True)
    model_fwd_total = pd.concat(model_fwd_list).reset_index(drop=True)
    dur[1] = time.time()
    qh, rh = dur.sum()// 3600, dur.sum() % 3600
    qm, rm = rh // 60, rh % 60
    print(f"Elapsed time: {int(qh)}h {int(qm)}min {int(rm)}s")
    return [df_fwd_total, model_fwd_total]
```
